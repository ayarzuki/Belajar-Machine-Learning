## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
plt.style.use('seaborn')
```

## Generate Data

```python
#create dataset
X, _ = make_blobs(n_samples=800, n_features=2, centers=3, cluster_std=1.9, shuffle=True, random_state=20)

#plot
plt.scatter(X[:, 0], X[:, 1], c='white', edgecolor='green')
plt.show()
```

## Model Data

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=20)
y_km = km.fit_predict(X)
```

```python
y_km
```

## Plot Cluster

```python
# Plot the 1 cluster
plt.scatter(X[y_km == 0, 0], X[y_km == 0, 1])
plt.show()
```

```python
# Plot the 3 clusters
plt.scatter(X[y_km == 0, 0], X[y_km == 0, 1], s=50, c='green', edgecolor='black', label='cluster 1')
plt.scatter(X[y_km == 1, 0], X[y_km == 1, 1], s=50, c='orange', edgecolor='black', label='cluster 2')
plt.scatter(X[y_km == 2, 0], X[y_km == 2, 1], s=50, c='blue', edgecolor='black', label='cluster 3')

#plot the centroids
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=30, c='red', edgecolor='black', label='centroids')

plt.legend()
plt.grid(False)
plt.show()
```

## Elbow Method

```python
# Menentukan key yang tepat
km.inertia_
```

inertia = total distance dari masing2 centroid

```python
distortions = []
for i in range(1, 11):
    km = KMeans(
        n_clusters=i, init='random',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=20
    )
    km.fit(X)
    distortions.append(km.inertia_)
    
# plot
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```
